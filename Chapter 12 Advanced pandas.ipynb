```python
N=15
```

```python
times = pd.date_range('2017 5 20 00:00',freq='1min',periods=N)
times
```

```python
df = pd.DataFrame({
    'time':times,
    'value':np.arange(N)
})
df
```

```python
df.set_index('time').resample('5min').count()
```

```python
df2 = pd.DataFrame({
    'time':times.repeat(3),
    'key':np.tile(['a','b','c'],N),
    'value':np.arange(N*3)
})
df2
```

```python
resampled = (df2.set_index('time').groupby(['key','time']).sum())
resampled
```
