```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
path = 'pydata-book-3rd-edition/datasets/movielens/'

unames =  ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table(path+'users.dat',header=None,sep='::',names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(path+'ratings.dat',header=None,sep='::',names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(path+'movies.dat', sep='::',header=None, names=mnames)
```

```python
users
```

```python
ratings
```

```python
movies
```

```python
data = pd.merge(pd.merge(users,ratings),movies)
data
```

```python
mean_ratings = data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')
mean_ratings
```

```python
data
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
active_titles = ratings_by_title[ratings_by_title>=250].index
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
top_female_ratings
```

```python
top_female_ratings['difference']=top_female_ratings['M']-top_female_ratings['F']
top_female_ratings
```

```python
top_female_ratings.sort_values(by='difference',ascending=False)
```
