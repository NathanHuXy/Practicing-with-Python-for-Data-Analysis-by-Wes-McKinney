# Remember:
___During the course of doing data analysis and modeling, a significant amount of time is spent on data preparation: loading, cleaning, transforming, and rearranging. Such tasks are often reported to take up 80% or more of an analystâ€™s time. ___

# Let's make a brief outline
- Missing data: `dropna()`
- replenish the missing data: `fillna()`
- do element-wise operation for a column: `map()`, pass a function or a dict.
- change index and columns: `rename(index=,columns=)` pass functions or dicts; or use `map()`,but it's not as convenient as `rename()`.
- binning: `obj=pd.cut(ages,bins)`,where `bins` is a list of all points. Use `pd.value_counts(obj)`,`obj.categories`.
- outliers: select them with logic and give them a value. example: `data[np.abs(data)>3]=np.sign(data)*3`
- rearange the order: `sampler = np.random.permutation(5)` to generate a new 5-emelemts order, then use `df.take(sampler)` to get the new dataframe.
- sampling: `df.sample(n=,replace)`
- dummy variables: `pd.get_dummies(df['key'],prefix='key')`. 
    - creat a big Dataframe, and write inside it(not speedy when big data)
    - creat a big ndarray, and write inside it (very speed when big data)
    - A usefule recipe is use `pd.get_dummies()` with `pd.cut()` at the same time. example: `pd.get_dummies(pd.cut(values,bins))`

```python
mnames = [
    'movie_id',
    'title',
    'genres'
]
movies = pd.read_table('movies.dat',sep='::',header=None,names=mnames,engine='python')
```

```python
movies
```

```python
all_genres=[]
for x in movies['genres']:
    all_genres.extend(x.split('|'))
#     print(set(x.split('|')))
#     print(all_genres)

genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies),len(genres)))
dummies = pd.DataFrame(zero_matrix,columns=genres)
dummies
```

```python
for i,gen in enumerate(movies['genres']):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i,indices]=1
```

```python
dummies
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic
```

```python
movies_windic.iloc[0]
```
