```python
%%time

nutrients = []

for rec in db:
    fnuts = pd.DataFrame(rec["nutrients"])
    fnuts['id']=rec['id']  #This line isn't included in author's code!!!!
    nutrients.append(fnuts)

nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
nutrients
```

```python
nutrients.duplicated().sum()
```

```python
nutrients = nutrients.drop_duplicates()
nutrients
```

```python
info = info.rename(
    columns = {
        'description':'food',
        'group':'fgroup'
    },
    copy=False
)
info
```

```python
nutrients = nutrients.rename(
    columns = {
        'description':'nutrient',
        'group':'nutgroup'
    },
    copy=False
)
nutrients
```

```python
ndata = pd.merge(nutrients,info,on='id',how='outer')
```

```python
ndata
```

```python
result = ndata.groupby(['nutrient','fgroup'])['value'].quantile(0.5)
result
```

```python
result.sort_values()[-30:].plot(kind='barh',figsize=(13,10))
```
