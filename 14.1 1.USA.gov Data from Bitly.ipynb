```python
cframe = frame[frame['a'].notnull()]
```

```python
cframe['os']=np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
```

```python
cframe['os'].head(10)
```

```python
by_tz_os = cframe.groupby(['tz','os'])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts
```

```python
indexer = agg_counts.sum(1).argsort()
indexer
```

```python
agg_counts
```

```python
indexer = agg_counts.sum(axis=1).argsort()
```

```python
indexer
```

```python
sns.barplot(x=indexer[:10].values, y=indexer[:10].index)
```

```python
count_subset = agg_counts.take(indexer[-10:])
count_subset
```

```python
agg_counts.sum(1).nlargest(10)
```

```python
count_subset
```

```python
count_subset = count_subset.stack()
```

```python
count_subset.name='total'
```

```python
count_subset = count_subset.reset_index()
```

```python
count_subset
```

```python
sns.barplot(x='total',
           y='tz',
           hue='os',
           data=count_subset)
```

```python
count_subset
```

```python
count_subset['rate'] = count_subset['total']/count_subset['total'].sum()
count_subset
```

```python
sns.barplot(x='rate',y='tz',hue='os',data=count_subset)
```

```python
count_subset = count_subset[count_subset['tz']!='']
count_subset
```

```python
f = pd.DataFrame({
    'key':list('abab'),
    'data':range(1,5)
})
f
```

```python
def norm(group):
    return group/group.sum()
f.groupby('key').apply(norm)
```

```python
def norm_total(group):
    group['normed_total'] = group['total']/group['total'].sum()
    return group
    
results = count_subset.groupby('tz').apply(norm_total)
```

```python
results
```

```python
sns.barplot(x='normed_total',y='tz',hue='os',data=results)
```
